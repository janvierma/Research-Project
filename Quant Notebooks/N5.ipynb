```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import yfinance as yf 
%matplotlib.inline 
```

```python
start = '2018-01-01'
end = '2021-12-31'
data = yf.download(['AAPL','MSFT','AMZN','GOOG'],start,end)
data
```

```python
df = data['Adj Close']
df
```

```python
df['MSFT'].head()
```

```python
plt.hist(df['MSFT'],bins=30)
plt.xlabel('price')
plt.ylabel('number of days observed')
plt.legend
```

```python
px.histogram(df['AMZN'],title='frequency distribution of returns of AAPL',template='plotly_dark',labels={"count":"num of days"})
```

```python
px.histogram(df['GOOG'],title='frequency distribution of prices of google',template='plotly_dark',labels={"value":"price","count":"frequency"})
```

```python
r = df['GOOG'].pct_change()[1:]
px.histogram(r,template='ggplot2')
```

Cumulative Histogram [Discrete estimated CDF]


```python
from tkinter.tix import Tree


px.histogram(r,template='plotly_dark',cumulative=True)
```
