Normal Distribution using mean returns and return standard deviation

```python
plt.hist( np.random.normal(amzn_mean,amzn_std,10000),bins=200)
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['Amazon Returns'])
```

```python
from turtle import color


plt.hist( np.random.normal(amzn_mean,amzn_std,10000),bins=75,color='cyan')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['Amazon Returns'])
plt.title('Normal Probability Distribution')
```

```python
#movingAVG = pd.rolling(amzn_adj,window=30) got an error, better to use the rolling function from pandas series, amzn_adj has close prices
movingAVG = amzn_adj.rolling(30).mean()
plt.plot(amzn_adj.index,amzn_adj.values)
plt.plot(movingAVG.index,movingAVG.values)
plt.ylabel('Price')
plt.xlabel('Time')
plt.legend(['AMZN','30-DAY MAVG'])
plt.title('Moving Average - 30 days')
```
